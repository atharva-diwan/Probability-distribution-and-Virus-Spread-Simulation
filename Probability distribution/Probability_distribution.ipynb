```
import matplotlib.pyplot as plt
import numpy as np
```

```
x=np.zeros(16)
```

```
j=3
for i in range(16):
  x[i]=j
  j=j+1

print(x.shape)
```

```
import random
y=np.zeros(16)
for i in range(1000):
  temp_sum=random.randint(1,6)+random.randint(1,6)+random.randint(1,6)
  for j in range(16):
    if temp_sum==x[j]:
      y[j]=y[j]+1
y=np.multiply(y,0.001)
```

```
from scipy.signal import savgol_filter
yhat = savgol_filter(y, 7, 3) # window size 5, polynomial order 3
plt.title("Probability distrbution curve of 3 dice rolled 1000 times")
plt.xlabel("Sum of the numbers on 3 dice")
plt.ylabel("probability of Sum of numbers to be X")
plt.plot(x,y)
plt.plot(x,yhat, color='red')
plt.show()
plt.savefig('plot1.jpg')
```

```
plt.bar(x, y, width=0.8)
plt.xlabel("Sum of the numbers on 3 dice")
plt.ylabel("probability of Sum of numbers to be X")
plt.show()
plt.savefig('plot2.png')
```
